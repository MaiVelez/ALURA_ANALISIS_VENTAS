### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos_tienda_1 = tienda["Precio"].sum()
ingresos_tienda_2 = tienda2["Precio"].sum()
ingresos_tienda_3 = tienda3["Precio"].sum()
ingresos_tienda_4 = tienda4["Precio"].sum()
print(f"Facturación T1 {ingresos_tienda_1}")
print(f"Facturación T2 {ingresos_tienda_2}")
print(f"Facturación T3 {ingresos_tienda_3}")
print(f"Facturación T4 {ingresos_tienda_4}")
```

```python
#Tienda 1 con mayor nivel de facturacion, por el contrario Tienda 4 tiene las ventas mas bajas.
```

```python
import matplotlib.pyplot as plt
facturacion_total = ingresos_tienda_1 + ingresos_tienda_2 + ingresos_tienda_3 + ingresos_tienda_4
print(f"El total de ingresos de todas las tiendas fue de:$ {facturacion_total}")
```

```python
plt.bar (x=["Tienda 1","Tienda 2","Tienda 3","Tienda 4"] ,height=(ingresos_tienda_1, ingresos_tienda_2, ingresos_tienda_3, ingresos_tienda_4), color ="blue")
```

# 2. Ventas por categoría

```python
datos_agrupados = tienda.groupby(["Categoría del Producto"])[["Precio"]].sum().sort_values("Precio",ascending=False)
datos_agrupados
```

```python
# Gran nivel de ventas sector de electrodomesticos, lamentablemente el sector de libros presenta varios retos en su estrategia de ventas.
```

```python
import matplotlib.pyplot as plt
datos_agrupados.plot.bar(color="blue")
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = tienda["Calificación"].mean().round(7)
print(f"La Calificación de la Empresa es de: {calificacion_tienda}")
```

# 4. Productos más y menos vendidos

```python
tienda.groupby(["Producto"])[["Precio"]].sum().sort_values("Precio", ascending=False)
```

```python
#El producto que mas se vende son los televisores, refrigeradores y celulaes.
#Los productos que menos se venden son la cuerda para saltar, Dinosaurio Rex y el cubo magico
```

# 5. Envío promedio por tienda

```python

envio_promedio_tienda_1 = tienda["Costo de envío"].mean().round(7)
envio_promedio_tienda_2 = tienda2["Costo de envío"].mean().round(7)
envio_promedio_tienda_3 = tienda3["Costo de envío"].mean().round(7)
envio_promedio_tienda_4 = tienda4["Costo de envío"].mean().round(7)
print(f"El envío promedio tienda 1 es de : {envio_promedio_tienda_1}")
print(f"El envío promedio tienda 2 es de : {envio_promedio_tienda_2}")
print(f"El envío promedio tienda 3 es de : {envio_promedio_tienda_3}")
print(f"El envío promedio tienda 4 es de : {envio_promedio_tienda_4}")
```

```python
# El promedio de envios es directamente proporcional a las ventas. Mayor costos de envio los tiene la tienda 1 por su buen nivel de ventas.
```
